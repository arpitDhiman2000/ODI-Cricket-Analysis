```python
import pandas as pd
import json
import numpy as np
```

```python
with open('./lines.json') as f:
    data = json.load(f)
    
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.index = np.arange(1, len(df_match)+1)
df_match.head(5)
```

```python
df_match.rename({'scorecard':'match_id'}, axis = 1, inplace = True)
df_match.head(5)
```

```python
df_match.to_csv('./Preprocessed Data/match_summary.csv', index=False)
```

```python
match_id_dict = {}

for index, row in  df_match.iterrows():
    
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
```

```python
df_batting = pd.read_excel('./Final Data/Book2.xlsx')
df_batting
```

```python
df_batting.drop(columns='Sr. No.')
```

```python
for i in range(0,876):
    text = str(df_batting.match[i])
    text = text.strip()
    df_batting.match[i] = text
```

```python
df_batting['dismissal'] = df_batting['dismissal'].fillna('')
df_batting
```

```python
# df_batting.dismissal.isna().sum()
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting
```

```python
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting
```

```python
# nan_value = df_batting['match_id'].isna().sum()
# nan_value
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('(c)', ''))
df_batting
```

```python
df_batting.to_csv('./Preprocessed Data/batting_summary.csv', index=False)
```
